Nå brukte jeg "random" pakken inne i "numpy" for å hente ut 2 tilfeldige tall mellom 0 og 1.
Du kan få et nesten vilkårlig mange tilfeldige tall

```python
a = np.random.random(1000)
print(len(a),a) #len() er for å sjekke hvor mange element som er inneholdt i variabelen a
```

så kan vi prøve å lage noe mer visuelt for å se om det faktisk er tilfeldig ved å bruke for eksempel
matplotlib, en modul som lager grafer for deg

```python
import matplotlib.pyplot as plt
plt.plot(a)
plt.show()
```

```python
np.mean(a) #finner gjennomsnittet av alle tallene inneholdt i variabelen a
```

her ser vi at snittet ligger ganske bra mot 0.5, midten av fordelingen.
vi kan lage flere tilfeldige tall for at snittet skal komme nærmere 0.5

```python
for i in range(9):
    print(np.mean(np.random.random(10**i)))
```
